```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('customers - customers.csv')
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:,[3,4]].values
X
```

```python
plt.scatter(X[...,0],X[...,1])
plt.xlabel('Total Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()         
```

```python
kmeans = KMeans(n_clusters=5,init='k-means++',random_state=0)
Y_kmeans = kmeans.fit_predict(X)
Y_kmeans
```

# AUDIENCE GROUPED ON BASIS OF BEHAVIOURAL SEGMENTAION 

```python
plt.scatter(X[Y_kmeans==0,0],X[Y_kmeans==0,1],label='Smart')
plt.scatter(X[Y_kmeans==1,0],X[Y_kmeans==1,1],label='Standard')
plt.scatter(X[Y_kmeans==2,0],X[Y_kmeans==2,1],label='Spendthrift')
plt.scatter(X[Y_kmeans==3,0],X[Y_kmeans==3,1],label='Careless')
plt.scatter(X[Y_kmeans==4,0],X[Y_kmeans==4,1],label='Sensible')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
